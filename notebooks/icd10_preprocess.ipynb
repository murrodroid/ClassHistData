```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('../datasets/ICD10codes.csv')
df
```

```python
def add_subcategory(df, target_col='ICD10', overcat_col='overcategory', output_col = 'subcategory'):
    df[output_col] = df.apply(
        lambda row: row[target_col][len(row[overcat_col]):],
        axis=1
    )
    return df

df = add_subcategory(df)
```

```python
def add_processed_desc(df, target_col='desc1', output_col = 'desc_processed'):
    df[output_col] = df.apply(
        lambda row: re.sub(r'[^a-z0-9 ]', '', row[target_col].lower()),
        axis=1
    )
    return df

df = add_processed_desc(df)
```
